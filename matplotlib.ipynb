# Matplotlib for visualisation

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

plt.figure(figsize=(5,5))
empid=np.arange(1, 11)
salary=np.arange(1, 11)
plt.bar(empid, salary, color='red', alpha=0.9)
plt.xlabel("Employee id")
plt.ylabel("Salary")
plt.show()
```
