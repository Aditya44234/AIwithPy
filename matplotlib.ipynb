```python
import matplotlib.pyplot as plt

# Data

plt.figure(figsize=(5,5))
years = ['2015', '2016', '2017', '2018', '2019', '2020']
revenue = [350, 400, 380, 410, 450, 500]

# Create the bar graph
colors = ['lightblue' if val < 450 else 'steelblue' for val in revenue]  # Color coding based on revenue
plt.bar(years, revenue, color=colors)

# Add numerical labels on top of each bar
for i, value in enumerate(revenue):
    plt.text(i, value + 10, str(value), ha='center', fontsize=9, color='black')

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Annotate significant event in 2020
plt.annotate('Product Launch', 
             xy=(5, 500), xytext=(4.5, 520), 
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Revenue (in millions)')
plt.title('Annual Revenue Trends')

# Display the graph
plt.tight_layout()
plt.show()
```

Grouped bargraph

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.figure(figsize=(6,6))
#Set Width of bar
barWidth = 0.25
#Set height of Bar
plt.title("Grouped Bar graph")
y1=[17,32,45,23]
y2=[43,6,19,23]
y3=[5,9,11,12]

#Set position of bar on X-axis
pos1=np.arange(len(y1))
pos2=[x+barWidth for x in pos1]
pos3=[x+barWidth for x in pos2]

# Make the plot 

plt.bar(pos1,y1,color='red',width=barWidth,label='Batting',alpha=0.5)
plt.bar(pos2,y2,color='green',width=barWidth,label='Batting',alpha=0.5)
plt.bar(pos3,y3,color='blue',width=barWidth,label='Feilding',alpha=0.5)
plt.xlabel("Players")
plt.ylabel("Scores")

plt.xticks([ i+barWidth for i  in range(len(y1))],['Virat','Rohit','Dhoni','Aditya'])
plt.legend()
plt.show()




```
