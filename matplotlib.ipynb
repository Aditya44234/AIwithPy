# Matplotlib for visualisation

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

plt.figure(figsize=(5,5))
empid=np.arange(1, 11)
salary=np.arange(1, 11)
plt.bar(empid, salary, color='red', alpha=0.5)
plt.xlabel("Employee id")
plt.ylabel("Salary")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

plt.figure(figsize=(6,5))
plt.title("Avg scores of all Subjects")
subj=["Python","DS","DBMS","OS","JAVA","System design"] #7
marks=[60,70,40,30,23,90]
plt.bar(subj,marks,color='red',alpha=0.4)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

plt.figure(figsize=(7,5))

plt.title("Annual revenue of company over the Years")
revn=[30,40,12,45,20,60]
year=[2015,2016,2017,2018,2019,2020]

plt.xlabel("Revenue")
plt.ylabel("Year")

plt.bar(year,revn,color='grey',alpha=0.5)
plt.show()
```
