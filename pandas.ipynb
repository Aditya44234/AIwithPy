```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dict={
    "name":['Aditya','Adarsh','Rohan','Abhiram'],
    "marks":[34,42,56,18],
    "section":['CSM','CSE','CSD','ECE']
}

df=pd.DataFrame(dict)
# df.to_csv('students.csv')
# df.to_excel('students.xlsx')


df.index=[1,2,3,4]
plt.figure(figsize=(6,3))
plt.barh(df['name'],df['marks'] ,color='grey')
plt.xlabel("Std Name")
plt.ylabel("Std Marks")
plt.title("Student marks")
```
